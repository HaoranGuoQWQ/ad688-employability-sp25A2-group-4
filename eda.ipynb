```python
import pandas as pd

df = pd.read_csv("lightcast_job_postings.csv")
```

2 Exploratory Data Analysis & Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create average salary column using SALARY_FROM and SALARY_TO
df['avg_salary'] = (df['SALARY_FROM'] + df['SALARY_TO']) / 2

plt.figure(figsize=(10, 6))
sns.histplot(df['avg_salary'].dropna(), bins=50, kde=True)
plt.title("Average Salary Distribution")
plt.xlabel("Average Salary")
plt.ylabel("Count")
plt.show()
```

```python
top_locations = df['STATE_NAME'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_locations.values, y=top_locations.index)
plt.title("Top 10 Job Postings by State")
plt.xlabel("Number of Postings")
plt.ylabel("State")
plt.show()
```
